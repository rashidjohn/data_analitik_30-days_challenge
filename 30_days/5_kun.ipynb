# Seaborn

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.DataFrame({
    "yosh": [23, 45, 31, 35, 40],
    "daromad": [3000, 6000, 5000, 5500, 7000],
    "jins": ["Erkak", "Ayol", "Erkak", "Ayol", "Erkak"]
})
```

scatterplot

| Parametr  | Ma’nosi                                       |
| --------- | --------------------------------------------- |
| `x`       | X o‘qidagi ustun                              |
| `y`       | Y o‘qidagi ustun                              |
| `hue`     | Rang bo‘yicha farqlash (kategoriya)           |
| `style`   | Belgilar turini farqlash (kategoriya)         |
| `size`    | Belgilar o‘lchamini farqlash                  |
| `palette` | Ranglar palitrasi (`deep`, `pastel`, `muted`) |
| `alpha`   | Shaffoflik (0–1)                              |


```python
sns.set_style("whitegrid")  # fonni bezatish
sns.scatterplot(x="yosh", y="daromad", hue="jins", data=data)

plt.title("Yosh va daromad orasidagi bog‘liqlik")
plt.xlabel("Yosh")
plt.ylabel("Daromad")
plt.show()
```

lineplot

```python
sns.lineplot(x="yosh", y="daromad", data=data)
plt.show()
```

barplot

Bu yerda "jins" bo‘yicha o‘rtacha daromad ko‘rsatiladi.

```python
sns.barplot(x="jins", y="daromad", data=data)
plt.show()
```

| Parametr    | Ma’nosi                                           |
| ----------- | ------------------------------------------------- |
| `x`, `y`    | Kategorik va sonli ustunlar                       |
| `data`      | DataFrame                                         |
| `hue`       | Rangga qarab yana bir kategoriyaga bo‘lish        |
| `estimator` | O‘rtacha (`mean`), median (`np.median`) va hokazo |
| `ci`        | Ishonch oralig‘i: `95` (default), `None`          |
| `palette`   | Ranglar to‘plami (`pastel`, `deep`, `Set2`, ...)  |

boxplot

Minimum, maksimum, median, kvartillarni ko‘rsatadi.

```python
sns.set_style("whitegrid")
sns.barplot(x="jins", y="daromad", data=data, palette="muted")

plt.title("Jins bo‘yicha o‘rtacha daromad")
plt.xlabel("Jins")
plt.ylabel("O‘rtacha daromad")
plt.show()
```

Har bir ustun jinslar bo‘yicha o‘rtacha daromadni ko‘rsatadi.

Bu yerda: Ayollar daromadi = (6000+5500)/2, Erkaklar daromadi = (3000+5000+7000)/3

**boxplot** - Box-and-Whisker diagramma

Sonli ma’lumotlar taqsimoti, median, kvartillar va g‘ayrioddiy qiymatlar (outliers) ni ko‘rsatadi.

| Parametr     | Ma’nosi                                  |
| ------------ | ---------------------------------------- |
| `x`, `y`     | Kategoriya va sonli ustun                |
| `data`       | DataFrame                                |
| `hue`        | Kategoriyalarni rang bilan ajratish      |
| `orient`     | `"v"` (vertical) yoki `"h"` (horizontal) |
| `palette`    | Ranglar (`Set2`, `pastel`, `muted`, ...) |
| `showfliers` | Outlier’larni ko‘rsatish: `True`/`False` |


```python
sns.set_style("whitegrid")
sns.boxplot(x="jins", y="daromad", data=data, palette="Set2")

plt.title("Jins bo‘yicha daromad taqsimoti")
plt.xlabel("Jins")
plt.ylabel("Daromad")
plt.show()
```

Har bir "jins" uchun daromadning taqsimoti ko‘rsatiladi.

Median: chiziq quti ichida.

Qutining balandligi → tarqalish.

Nuqtalar: outlier bo‘lishi mumkin.

**histplot** - histogramma (taqsimotni ko‘rsatadi)

Qancha qiymat qanday oraliqda joylashganini ko‘rsatadi.

| Parametr  | Ma’nosi                                              |
| --------- | ---------------------------------------------------- |
| `data`    | Pandas DataFrame                                     |
| `x`       | Ko‘rsatiladigan sonli ustun                          |
| `bins`    | Nechta bo‘linma (bin) bo‘lsin                        |
| `kde`     | Silliqlashtirilgan chiziqni chizish (`True`/`False`) |
| `hue`     | Ranglar orqali farqlash                              |
| `element` | Bar uslubi (`bars`, `step`, `poly`)                  |
| `stat`    | `count`, `density`, `percent`                        |

```python
data1 = pd.DataFrame({
    "baholar": [55, 60, 65, 70, 75, 80, 85, 85, 90, 90, 95, 100, 100, 100]
})

sns.set_style("whitegrid")
sns.histplot(data=data1, x="baholar", bins=5, kde=True, color='skyblue')

plt.title("Baholar tarqalishi")
plt.xlabel("Baholar")
plt.ylabel("Talabalar soni")
plt.show()
```

bins=5 — baholar 5 oraliqqa bo‘linadi

kde=True — silliq chiziq chiziladi (taqsimot zichligi)

Grafik sizga qaysi baholar eng ko‘p uchrayotganini ko‘rsatadi

**heatmap** -  issiqlik xaritasi (matritsalarni vizualizatsiya qilish)

| Parametr     | Ma’nosi                                                |
| ------------ | ------------------------------------------------------ |
| `data`       | 2D matritsa (`DataFrame` yoki `numpy array`)           |
| `annot`      | Jadvalga sonlarni yozish (`True`/`False`)              |
| `fmt`        | Format (masalan, `.2f`, `%d`)                          |
| `cmap`       | Ranglar xaritasi (`coolwarm`, `viridis`, `Blues`, ...) |
| `linewidths` | Kataklar oralig‘i                                      |
| `linecolor`  | Kataklar chizig‘i rangi                                |


Ustunlar orasidagi bog‘liqlik (korrelyatsiya) darajasini rang orqali ko‘rsatadi.

```python
tips = sns.load_dataset("tips")

correlation = tips.corr(numeric_only=True)

sns.heatmap(correlation, annot=True, cmap="YlGnBu", fmt=".2f")
plt.title("Tips datasetda ustunlar orasidagi korrelyatsiya")
plt.show()
```

Har bir katak -1 dan +1 gacha qiymatga ega

+1 → kuchli ijobiy bog‘liqlik (birgalikda oshadi)

-1 → kuchli salbiy bog‘liqlik (birga kamayadi)

0 → umuman bog‘liqlik yo‘q

Ranglar orqali tezda qaysi ustunlar bir-biriga bog‘liq ekanini ko‘rish mumkin

Grafikni bezash (stil, rang, o‘lcham)

```python
sns.set_style("darkgrid")  # yoki 'white', 'ticks', 'dark'
sns.set_palette("pastel")  # Ranglar palitrasi

sns.barplot(x="jins", y="daromad", data=data)
plt.title("Jins bo‘yicha daromad")
plt.xlabel("Jins")
plt.ylabel("Daromad")
plt.show()
```

Ichki ma’lumotlar to‘plami bilan ishlash

**pairplot** - Juft grafiklar majmuasi

| Parametr    | Ma’nosi                                                        |
| ----------- | -------------------------------------------------------------- |
| `data`      | DataFrame                                                      |
| `hue`       | Rang orqali ajratiladigan ustun (kategoriya)                   |
| `diag_kind` | Diagonal qismda nima bo‘lsin: `"auto"`, `"hist"`, `"kde"`      |
| `kind`      | Pastki uchburchakdagi grafiklar turi: `"scatter"` yoki `"reg"` |
| `palette`   | Rang to‘plami                                                  |
| `markers`   | Belgilar (marker) turi (masalan: `["o", "s"]`)                 |


```python
iris = sns.load_dataset("iris")

sns.pairplot(data=iris, hue="species", diag_kind="kde", palette="husl")

plt.suptitle("Iris datasetda ustunlar juftliklari", y=1.02)
plt.show()
```

Har bir juft ustun uchun scatterplot

Diagonal chiziqlarda — har bir ustun uchun taqsimot (histogram yoki KDE)

hue="species" → ranglar orqali species ustuni bo‘yicha ajratilgan

Ko‘rish mumkin: qaysi turning (species) petal_length va petal_width o‘rtasida kuchli farq bor

```python
tips = sns.load_dataset("tips")

sns.pairplot(data=tips, vars=["total_bill", "tip", "size"], hue="sex", palette="Set2")
plt.suptitle("Tips datasetda juft ustunlar", y=1.02)
plt.show()
```

| Parametr  | Ma’nosi                                        |
| --------- | ---------------------------------------------- |
| `x`, `y`  | X va Y o‘qlaridagi ustunlar                    |
| `data`    | DataFrame                                      |
| `hue`     | Rang bilan ajratish (kategoriya bo‘yicha)      |
| `col`     | Grafikni ustun bo‘yicha ajratish               |
| `row`     | Grafikni qatorda ajratish                      |
| `palette` | Ranglar to‘plami                               |
| `markers` | Belgilar ko‘rinishi                            |
| `ci`      | Ishonch oralig‘i (`None` qilinsa chizilmaydi)  |
| `fit_reg` | Regressiya chizig‘ini chizish (`True`/`False`) |


Eslatma: lmplot() — bu figure-level plot, ya’ni plt.show() dan keyin facetgrid obyekt hosil qiladi.

```python
# Ma'lumot
tips = sns.load_dataset("tips")

# Regressiya grafigi
sns.lmplot(x="total_bill", y="tip", data=tips)

plt.title("Hisob va choychaqa orasidagi regressiya")
plt.show()
```

Bu orqali erkak va ayol mijozlarning regressiya chizig‘ini alohida ko‘rish mumkin

```python
sns.lmplot(x="total_bill", y="tip", hue="sex", data=tips, palette="Set1")
```

Linear regression uchun vizual tekshiruv
