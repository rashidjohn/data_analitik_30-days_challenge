Slicing

```python
# string ni kesib olish
matn = 'Hello world!'
print(matn[1:7]) # 1 dan 7-index gacha(7 kirmaydi)
print(matn[:3]) # boshidan 3-index gacha(3 kirmaydi)
print(matn[5:]) # 5-index dan oxirigacha
print(matn[::-1]) # matnni oxiridan boshiga(teskari tartibda)
print(matn[::2]) # boshidan oxirigacha 2 qadam bilan kesib olish
```

```python
i = 1
while i < 6:
  print(i)
  i += 1
else:
  print("i is no longer less than 6")
```
