```python
import pandas as pd
import numpy as np
```

```python
df_original = pd.DataFrame({
    'Name': ['Ali', 'Vali', 'Gulnora'],
    'Age': [27, 32, 45],
    'Salary': [600, 1200, 4000],
    'Joined_Date': ['2015-06-01', '2010-08-10', '2005-09-12']
})
df_original
```

```python
df_original['Joined_Date'] = pd.to_datetime(df_original['Joined_Date'])
df_original['Experience_Years'] = 2025 - df_original['Joined_Date'].dt.year
df_original['Log_Salary'] = np.log1p(df_original['Salary'])
df_original
```

```python
df_gender = pd.DataFrame({
    'Gender': ['Male', 'Female', 'Female', 'Male']
})
df_gender
```

```python
df_gender['Gender_Label'] = df_gender['Gender'].map({'Male': 1, 'Female': 0})
df_gender
```

```python
df_encoded = pd.get_dummies(df_gender, columns=['Gender'], drop_first=True)
df_encoded
```

```python
df = pd.concat([df_original, df_gender], axis=1)
df
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

scaler = StandardScaler()
df['Salary_scaled'] = scaler.fit_transform(df[['Salary']])

print(df)
```

```python
df.drop(['Name','Gender'], axis=1).corrwith(df['Salary']).abs().sort_values(ascending=False)
```
