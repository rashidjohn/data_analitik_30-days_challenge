```python
import pandas as pd
import numpy as np
```

dataframe yaratamiz

```python
df1 = pd.DataFrame({
    'id': [1, 2, 3],
    'ism': ['Ali', 'Vali', 'Hasan']
})

df2 = pd.DataFrame({
    'id': [1, 2, 4],
    'yosh': [23, 25, 30]
})
```

```python
df1
```

```python
df2
```

merge() — JOIN orqali birlashtirish

bir xil id ga ega bo'lganlar birlashadi

```python
pd.merge(df1, df2, on='id')
```

**how** parametridan foydalanish

**merge()** bilan to‘rt xil **how** mavjud:

| `how` qiymati | Ma'nosi                          | SQL'dagi mosligi |
| ------------- | -------------------------------- | ---------------- |
| `'inner'`     | Umumiy qism (default)            | INNER JOIN       |
| `'left'`      | Chap jadvalni saqlaydi           | LEFT JOIN        |
| `'right'`     | O'ng jadvalni saqlaydi           | RIGHT JOIN       |
| `'outer'`     | Har ikkala jadvalni to‘liq oladi | FULL OUTER JOIN  |


```python
pd.merge(df1, df2, on='id', how='left')   # df1 asos
```

```python
pd.merge(df1, df2, on='id', how='right')  # df2 asos
```

```python
pd.merge(df1, df2, on='id', how='outer')  # ikkala tomon
```

**concat()** — ustun/qatorni birlashtirish

Qator bo‘yicha (index bo‘yicha) birlashtirish:

```python
df3 = pd.DataFrame({'ism': ['Aziz'], 'id': [4]})
df3
```

```python
new_df = pd.concat([df1, df3], ignore_index=True)
new_df
```

Ustun bo‘yicha birlashtirish:

```python
df4 = pd.DataFrame({'yosh': [20, 22, 24]})
df4
```

```python
new_df = pd.concat([df1, df4], axis=1)
new_df
```

**join()** metodi indesklar bo'yicha birlashtirish

```python
df1 = df1.set_index('id')
df1
```

```python
df2 = df2.set_index('id')
df2
```

```python
df1.join(df2, how='inner')
```

| Holat                    | Foydalaniladigan usul                     |
| ------------------------ | ----------------------------------------- |
| SQLdagi JOIN'ni qilish   | `merge()`                                 |
| Faqat qator qo‘shish     | `concat()` yoki `append()` (yangi versiyada mavjud emas)               |
| Fayllarni birlashtirish  | `concat()` (csv fayllar ro‘yxatini o‘qib) |
| Indeks bo‘yicha qo‘shish | `join()`                                  |


O'rganilgan bilimlar

| Usul       | Qachon ishlatiladi                              |
| ---------- | ----------------------------------------------- |
| `merge()`  | Kalit ustunlar bo‘yicha qo‘shish                |
| `concat()` | Jadval ustunlari yoki qatorlarini birlashtirish |
| `join()`   | Indekslar bo‘yicha birlashtirish                |
| `append()` | Qator qo‘shish (eskirgan)                       |
